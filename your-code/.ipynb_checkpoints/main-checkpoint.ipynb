#### What's the  average amount of money spent by country

```python
avam_coun={}

for i in set(orders.Country):
    avam_coun[i] = orders.amount_spent[orders.Country==i].mean()
avam_coun
```

#### What was the average amount spent per year ?

```python
avam_year={}

for i in set(orders.year):
    avam_year[i] = orders.amount_spent[orders.year==i].mean()
avam_year
```
